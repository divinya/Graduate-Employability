```python
import matplotlib.pyplot as plt

# Replace missing values in the salary column with a placeholder value (e.g., -1)
cell_df['salary'] = cell_df['salary'].fillna(0)

# Plot boxplot for the salary column
plt.figure(figsize=(8, 6))
plt.boxplot(cell_df['salary'], showmeans=True)  # Show means to have more points on the plot
plt.scatter([1]*len(cell_df['salary']), cell_df['salary'], alpha=0.3, color='black')  # Add individual data points
plt.title('Distribution of Salary')
plt.xlabel('Salary')
plt.ylabel('Values')
plt.grid(True)

# Set specific tick marks on the y-axis
plt.yticks(range(200000, 1000000, 100000))

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Aggregate counts of 'status' by 'sex'
status_counts = cell_df.groupby(['sex', 'status']).size().unstack()

# Plot stacked bar plot
ax = status_counts.plot(kind='bar', stacked=True, figsize=(10, 6))

# Add title and labels
plt.title('Distribution of Status by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

# Add percentage labels on each segment
for patch in ax.patches:
    width, height = patch.get_width(), patch.get_height()
    x, y = patch.get_xy()
    if height > 0:  # To avoid zero-height segments
        ax.text(x + width/2, y + height/2, f'{height/status_counts.sum().sum()*100:.1f}%', ha='center', va='center')

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot stacked bar plot for 'status' vs 'work_xp'
plt.figure(figsize=(10, 6))
sns.countplot(x='work_xp', hue='status', data=cell_df)

# Add title and labels
plt.title('Distribution of Status by Work Experience')
plt.xlabel('Work Experience')
plt.ylabel('Count')

# Add legend
plt.legend(title='Status')

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the bins for degree percentage ranges
degree_bins = [0, 50, 60, 70, 80, 90, 100]

# Create a new column to categorize degree percentages into bins
cell_df['degree_pct_bins'] = pd.cut(cell_df['degree_pct'], bins=degree_bins, right=False)

# Aggregate counts of 'status' by 'degree_pct_bins'
status_counts = cell_df.groupby(['degree_pct_bins', 'status']).size().unstack()

# Plot stacked bar plot
ax = status_counts.plot(kind='bar', stacked=True, figsize=(10, 6))

# Add title and labels
plt.title('Distribution of Status by Degree Percentage')
plt.xlabel('Academic Peeformance')
plt.ylabel('Count')

# Add percentage labels on each segment
for patch in ax.patches:
    width, height = patch.get_width(), patch.get_height()
    x, y = patch.get_xy()
    if height > 0:  # To avoid zero-height segments
        ax.text(x + width/2, y + height/2, f'{height/status_counts.sum().sum()*100:.1f}%', ha='center', va='center')

# Show plot
plt.show()
```
